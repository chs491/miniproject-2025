```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# ----------------------------------------
# 1. ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
# ----------------------------------------
df = pd.read_csv("electricity_data.csv")  # ì „ë ¥ëŸ‰ ë°ì´í„° CSV
values = df['ì „ë ¥ëŸ‰'].values.reshape(-1, 1)

# ë°ì´í„° ì •ê·œí™”
scaler = MinMaxScaler()
scaled = scaler.fit_transform(values)

# ----------------------------------------
# 2. ì‹œê³„ì—´ ë°ì´í„°ì…‹ ë§Œë“¤ê¸°
# ----------------------------------------
def create_dataset(data, time_steps=10):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:(i + time_steps), 0])
        y.append(data[i + time_steps, 0])
    return np.array(X), np.array(y)

X, y = create_dataset(scaled, time_steps=10)
X = X.reshape((X.shape[0], X.shape[1], 1))

# ----------------------------------------
# 3. LSTM ëª¨ë¸ ì •ì˜
# ----------------------------------------
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(10, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# ----------------------------------------
# 4. ëª¨ë¸ í•™ìŠµ
# ----------------------------------------
model.fit(X, y, epochs=50, batch_size=16, verbose=1)

# ----------------------------------------
# 5. ëª¨ë¸ ì €ì¥
# ----------------------------------------
model.save("electricity_lstm_model.h5")
print("âœ… electricity_lstm_model.h5 íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
import numpy as np
import pandas as pd
import pymysql
from sqlalchemy import create_engine
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
import os

# ----------------------------------------
# 1. MySQL DB ì—°ê²°
# ----------------------------------------
DB_HOST = "localhost"
DB_USER = "root"
DB_PASSWORD = "12345"
DB_NAME = "electric_db"
DB_PORT = 3306

# SQLAlchemy ì—”ì§„ ìƒì„±
try:
    engine = create_engine(f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}")
    # ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
    query = "SELECT ì‹œê°„, ì „ë ¥ëŸ‰ FROM ì „ë ¥ëŸ‰ ORDER BY ì‹œê°„ ASC"
    df = pd.read_sql(query, engine)
    print("ë°ì´í„° ë¡œë“œ ì™„ë£Œ")
except Exception as e:
    print(f"ë°ì´í„°ë² ì´ìŠ¤ ì—°ê²° ë˜ëŠ” ì¿¼ë¦¬ ì˜¤ë¥˜: {e}")
    exit()

# ì»¬ëŸ¼ ì´ë¦„ í™•ì¸
print("ì»¬ëŸ¼ ì´ë¦„:", df.columns.tolist())
power_column = 'ì „ë ¥ëŸ‰'  # ì‹¤ì œ ì»¬ëŸ¼ ì´ë¦„ìœ¼ë¡œ ë³€ê²½ í•„ìš”, ì˜ˆ: 'power_usage'
if power_column not in df.columns:
    print(f"ì˜¤ë¥˜: '{power_column}' ì»¬ëŸ¼ì´ ì—†ìŠµë‹ˆë‹¤. ì‚¬ìš© ê°€ëŠ¥í•œ ì»¬ëŸ¼: {df.columns.tolist()}")
    exit()

# ê²°ì¸¡ê°’ í™•ì¸ ë° ì²˜ë¦¬
if df[power_column].isnull().any():
    print(f"ê²½ê³ : '{power_column}' ì»¬ëŸ¼ì— ê²°ì¸¡ê°’ì´ ìˆìŠµë‹ˆë‹¤. ê²°ì¸¡ê°’ ì œê±° ì¤‘...")
    df = df.dropna()
    if df.empty:
        print("ì˜¤ë¥˜: ê²°ì¸¡ê°’ ì œê±° í›„ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
        exit()

# ë°ì´í„° ìœ íš¨ì„± í™•ì¸ (ìˆ«ìí˜•ì¸ì§€ í™•ì¸)
df[power_column] = pd.to_numeric(df[power_column], errors='coerce')
if df[power_column].isnull().any():
    print(f"ê²½ê³ : '{power_column}' ì»¬ëŸ¼ì— ìˆ«ìê°€ ì•„ë‹Œ ê°’ì´ í¬í•¨ë˜ì–´ ìˆìŠµë‹ˆë‹¤. ê²°ì¸¡ê°’ ì œê±° ì¤‘...")
    df = df.dropna()
    if df.empty:
        print("ì˜¤ë¥˜: ë¹„ìˆ«ì ê°’ ì²˜ë¦¬ í›„ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
        exit()

# ----------------------------------------
# 2. ë°ì´í„° ì „ì²˜ë¦¬
# ----------------------------------------
values = df[power_column].values.reshape(-1, 1)
scaler = MinMaxScaler()
scaled = scaler.fit_transform(values)
print(f"Scaled data length: {len(scaled)}")

# ----------------------------------------
# 3. ì‹œê³„ì—´ ë°ì´í„°ì…‹ ë§Œë“¤ê¸°
# ----------------------------------------
def create_dataset(data, time_steps=10):
    if len(data) <= time_steps:
        raise ValueError(f"ë°ì´í„° ê¸¸ì´({len(data)})ê°€ time_steps({time_steps})ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤.")
    
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:(i + time_steps), 0])
        y.append(data[i + time_steps, 0])
    X = np.array(X)
    y = np.array(y)
    
    if X.size == 0:
        raise ValueError("ìƒì„±ëœ X ë°°ì—´ì´ ë¹„ì–´ ìˆìŠµë‹ˆë‹¤.")
    
    return X, y

# ë°ì´í„° í¬ê¸° í™•ì¸
time_steps = 3
if len(scaled) <= time_steps:
    print(f"ì˜¤ë¥˜: ë°ì´í„° ê¸¸ì´({len(scaled)})ê°€ time_steps({time_steps})ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤.")
    print("í•´ê²° ë°©ë²•: ë” ë§ì€ ë°ì´í„°ë¥¼ ì¤€ë¹„í•˜ê±°ë‚˜ time_steps ê°’ì„ ì¤„ì´ì„¸ìš”.")
    exit()

# ë°ì´í„°ì…‹ ìƒì„±
try:
    X, y = create_dataset(scaled, time_steps=time_steps)
except ValueError as e:
    print(f"ë°ì´í„°ì…‹ ìƒì„± ì˜¤ë¥˜: {e}")
    exit()

# X ì°¨ì› í™•ì¸ ë° reshape
print(f"X shape before reshape: {X.shape}")
if X.size > 0 and len(X.shape) == 2:
    X = X.reshape((X.shape[0], X.shape[1], 1))
    print(f"X shape after reshape: {X.shape}")
else:
    print(f"ì˜¤ë¥˜: Xê°€ ë¹„ì–´ ìˆê±°ë‚˜ 2ì°¨ì› ë°°ì—´ì´ ì•„ë‹™ë‹ˆë‹¤. X shape: {X.shape}")
    print("í•´ê²° ë°©ë²•:")
    print("1. ë°ì´í„° ê¸¸ì´ë¥¼ í™•ì¸í•˜ì„¸ìš” (len(scaled) > time_steps).")
    print("2. create_dataset í•¨ìˆ˜ê°€ ì˜¬ë°”ë¥¸ ì°¨ì›ì˜ ë°ì´í„°ë¥¼ ë°˜í™˜í•˜ëŠ”ì§€ í™•ì¸í•˜ì„¸ìš”.")
    exit()

# ----------------------------------------
# 4. LSTM ëª¨ë¸ ì •ì˜
# ----------------------------------------
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(time_steps, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# ----------------------------------------
# 5. ëª¨ë¸ í•™ìŠµ
# ----------------------------------------
if X.shape[0] == 0 or y.shape[0] == 0:
    print("ì˜¤ë¥˜: í•™ìŠµ ë°ì´í„°(X ë˜ëŠ” y)ê°€ ë¹„ì–´ ìˆìŠµë‹ˆë‹¤.")
    exit()

model.fit(X, y, epochs=50, batch_size=16, verbose=1)

# ----------------------------------------
# 6. ëª¨ë¸ ì €ì¥
# ----------------------------------------
model_dir = "models"
os.makedirs(model_dir, exist_ok=True)
model_path = os.path.join(model_dir, "electricity_lstm_model.h5")
model.save(model_path)
print(f"âœ… {model_path} íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
